##¿Qué quiere decir que un problema **NP-Hard**? 

Que el algoritmo que resuelve el problema y entrega una solución requiere un tiempo exponencial en el peor de los casos. 

Debido a su naturaleza intrínsicamente compleja, no es posible resolverlos mediante ningún algoritmo rápido, es decir, algún algoritmo con tiempo de ejecución polinomial o menor.

Un problema es **NP-Hard** si es al menos tan difícil como todos los problemas con soluciones fácilmente reconocibles, refiriéndose a problemas donde se aplica la fuerza bruta.


##¿Qué quiere decir que $P \neq NP$ ?

Es una conjetura que declara que los problemas $P$ y $NP$ son distintos.  

* $P$ puede ser resuelto en tiempo polinomial.

* $NP$ no siempre puede ser resuelto en tiempo polinomial, solo puede ser **comprobado** en un tiempo polinomial.

Esto implica una diferencia en el concepto lógico al que ambas partes de la conjetura apelan.

##¿Cómo se prueba  que un problema es NP-Hard usando **reducciones**?

Si encontramos un problema **A** y queremos probar que es un problema **NP-Hard**, no serviría encontrar una solución exponencial, pues eso no significaría que no puede tener solución polinomial. 

Si queremos probar que el problema **A** es **NP-Hard** debemos seguir los siguientes pasos:

1. Encontrar un problema **B** que se sepa que es un **NP-Hard** y sea similar a nuestro problema **A**. Hay problemas **NP-Hard** conocidos.

2. Probar que el problema **B**, que es un **NP-Hard** conocido puede reducirse al problema **A** para obtener la solución.

Como **A** es un problema al que **B** puede reducirse y siendo **B** un algoritmo **Np-Hard**, entonces **A** también lo es.

##**Knapsack Problem :**

Este problema requiere de entrada $2n + 1$ números en la forma de: un conjunto de $n$ valores de elementos $v =v_1, v_2, ..., v_n$, un conjunto de $n$ pesos de elementos $w = w_1, w_2, ... , w_n$ y una capacidad de mochila $W$.

Busca calcular el subconjunto $S ⊆ {1,2,...,n}$ de elementos con suma máxima posible y cuyo tamaño total no sobrepase la capacidad $W$ de la mochila.

Es decir, la salida debe cumplir:

$ \displaystyle \sum_{i ϵ S}^{n ϵ S} v_i >> \sum_{i ϵ M}^{n ϵ M} v_i$ $𝝠$ $ \displaystyle \sum_{i ϵ S}^{n ϵ S} w_i \leq W$

> $M =$ subconjunto de $w$ distinto de $S$

$Input:$ $v$ valores, $w$ tamaños, $W$ capacidad.

$Output:$ $S$.

En vez de pesos $w$ pueden ser tamaños $s$, entonces la capacidad no sería un peso $W$, correspondería un tamaño $S$ y la salida tendría otra nomenclatura.

##**Knapsack Problem :**

La tarea es encontrar el máximo precio que podemos obtener de lo que podemos guardar en una *mochila* con una **capacidad de peso máxima** dada, es decir, maximizar las ganancias al vender los elementos de la *mochila* teniendo el precio y peso de cada *artículo* guardado.

Para resolver este problema se propone usar un método de **programación dinámica**. Donde antes de resolver el máximo precio de la mochila con capacidad $W$ y considerando $i$ artículos, debemos saber el máximo precio de la mochila con la misma capacidad que considera hasta $i-1$ artículos y la mochilas que consideran hasta $i$ articulo cuyo peso es menor a $W$.

###*Estructura del algoritmo*

Para considerar todos los subconjuntos de elementos, existe: un caso base y dos casos para cada iteración. Tendremos que usar una matriz con $n$ filas "$i$" que corresponden a cada artículo y con $W$ columnas que corresponden a las capacidades "$wt$" de distintas mochilas.

* **caso base** 
> Iteración $0$ o el peso máximo será $0$.

* **casos para cada iteración**:
> El *artículo* $i$ considerado supera por sí solo la capacidad $wt$ de la **mochila**, se considera como máximo valor para esa capacidad el máximo valor de la **mochila** con esa capacidad $wt$ que considera hasta el *articulo* $i-1$.

  > El *artículo* $i$ considerado no supera por sí solo la capacidad $wt$ de la **mochila**, en cuyo caso debemos escoger el mayor precio entre: La **mochila** que puede contener hasta el artículo $i-1$ **o** Una *mochila* cuyo peso es el de la **mochila** menos el peso del artículo a agregar que considera hasta el artículo $i-1$ más el valor del artículo $i$, es decir, el máximo precio de una mochila donde no se ha considerado el artículo $i$ más el precio del artículo $i$.

El algoritmo calcula el valor máximo para cada capacidad menor a la de la **mochila** buscada intentando agregar los artículos uno por uno hasta llegar al peso de la mochila.

$m =$ matriz

$i =$ clave actual

$p =$ peso actual

$p[i] =$ peso del i-ésimo artículo

$m[i] =$ valor del i-ésimo artículo

###*Tiempo de ejecución*

####Teorema *Complejidad Temporal*

*El algoritmo `Knapsack Problem` implementado a traves de **Bottom-up** tiene un **tiempo de ejecución** de* $O(n*W)$.

####Prueba del teorema

Para la resolución de este problema tenemos 2 bucles anidados, usados para recorrer los subproblemas a resolver antes de ser capaz de resolver el problema real (refiriéndose al aquellos problemas con menor tamaño al pedido).

Cada problema tiene en cuenta que tenemos el valor óptimo de sus subproblemas (el problema para cualquier mochila que soporta menos peso o artículos). Y con ese dato, dentro del bucle chico se resuelve el problema a todos los pesos posibles sacando el valor otpimo de cada uno al agregar el artículo $i$, dentro de este bucle se tiene un tiempo de ejecución $W \rightarrow O(W)$. 

Para el valor óptimo del tamaño pedido se deben resolver los subproblemas al agregar cada articulo $i=0$ **to** $n-1$. Provocando que se resuelvan los $W$ subproblemas al agregar cada artículo calculando su valor optimo con cada peso posible un total de $n$ veces, es decir, resolver $(n-1)*W$ subproblemas antes de resolver el problema real al agregar el ultimo articulo.

El problema real también tiene un tiempo de ejecución $W \rightarrow O(W)$.

Entonces se trata de resolver $n$ veces un problema con tiempo de ejecución $O(W)$. Por eso el algoritmo **bottom-up** para resolver `KnapSack Problem` es de $n * W → O(n*W)$.
