# **1. Problema de ordenamiento de datos**

**Entrada:** El programa recibe una secuencia de $n$ números arreglo en un orden aleatoria de la manera $[a_1, a_2, ..., a_n,]$.

**Salida:** El programa retorna la secuencia de $n$ números en un orden establecido de la manera $[a_1' \leq a_2'\leq ...,\leq a_n']$.

# **2. Insertion Sort**

## **2.1. Descripción del algoritmo**

Insertion Sort es uno de los algoritmos que solucionan el problema de ordenamiento de una forma sencilla.

El algoritmo recibe de entrada una lista(o arreglo) $arr$ con una secuencia de $n$ números en cualquier orden. 

Luego los números dentro de la lista(o arreglo) se ordenan de mayor a menor o de menor a mayor dependiendo del programador. En este caso describiremos el algoritmo al ordenar de menor a mayor

1.   Se recorrerá todo el arreglo
2.   En cada interacción i, se compara el $arr[i]$ con $arr[i - 1]$. 
Si $arr[i] < arr[i - 1]$ entonces se cambian de lugar y el proceso se repetirá mientras se cumpla esta condición.
3.   Luego de todas las interacciones, el arreglo debería de quedar ordenado de menor a mayor.
4.   El algoritmo retorna el arreglo y la cantidad de interacciones en el algoritmo.

## **2.2. Código**

```python
def insertionSort(arr, verbose=False): 
    n = len(arr)
    cont = 0 #contador de comparaciones

    if verbose == True: print(f"Arreglo Inicial = {arr}\n")

    # Se recorre el arreglo
    for i in range(1,n):
        
        if verbose == True: print("Pasada ",i)

        aux = arr[i] # Se guarda el i-esimo elemento que podria cambiar de posicion
        j = i - 1

        cont += 1
        
        if verbose == True: 
          print(f"{arr}")
          print(f"Comparación: arr[{j+1}] < arr[{j}] ===> {aux} < {arr[j]}")
          if aux >= arr[j]: print("No hubo cambios")
          
        # Se compara dentro del arreglo
        while j > -1 and aux < arr[j] :# Si el elemento es menor al anterior elemento, se intercambian
            arr[j+1] = arr[j]
            j -= 1
            if j==-1: cont -=1
            if verbose == True: print(f"Arreglo después del cambio {arr}")
            if verbose == True and j!=-1: print(f"Comparación: aux < arr[{j}] ===> {aux} < {arr[j]}")
            cont+= 1
        
        arr[j+1] = aux
        if verbose == True: print(f"Al finalizar pasada: {arr}\n")
    
    if verbose == True:
      for i in range(0,n-1):
          if arr[i] > arr[i+1]: 
              print("\nEl arreglo no se ordenó correctamente")
              break
          if i == n-2: print("\nEl arreglo se ordenó correctamente")
      print(f"Arreglo Final = {arr}")
      

    return cont

# Ejemplo
array = [1,8,3,7,9]
print("Datos de entrada: ",array)
counter = insertionSort(array)
print("Datos de salida: ",array)
print("# comparaciones: ", counter)
```

## **2.3. Ejemplo**

Por ejemplo, consideremos el siguiente arreglo

**a** = $[6,2,9,3]$

En la Primera pasada se compara si $2 < 6$, como el $2$ es menor a $6$, el $6$ ocupara la posición a su derecha y el $2$ quedara como auxiliar

**a** = $[6,6,9,3]$        `aux = 2`

Como no hay números antes del $6$, el *aux = 2* tomara su lugar

**a** = $[2,6,9,3]$

En la Segunda pasada se compara si $9 < 6$, como el $9$ no es menor a $6$, el arreglo permanecerá sin cambios

**a** = $[2,6,9,3]$

En la Tercera pasada se compara si $0 < 9$, como el $3$ es menor a $9$, el $9$ ocupara la posición a su derecha y el $3$ quedara como auxiliar

**a** = $[2,6,9,9]$        `aux = 3`

Ahora se compara si $3 < 6$ (el auxiliar < el elemento anterior a 7), como el $3$ es menor a $6$, el $6$ ocupará la posición a su derecha y el $0$ permanecerá como auxiliar

**a** = $[2,6,6,9]$        `aux = 3`

Ahora se compara si $3 < 2$ (el auxiliar < el elemento anterior a 6), como el $3$ no es menor a $2$, el $3$ ocupara la posición a la derecha del $2$

**a** = $[2,3,6,9]$

Al final de cada pasada, el arreglo ya tiene los ordenados hasta el elemento en la posición del número de la pasada en que se encuentra

# **3. Tiempo de ejecución**

### **Teorema (Tiempo de ejecución).**

*El algoritmo **Insertion Sort** tiene un **tiempo de ejecución de** $O(n^2)$ en el peor caso.* 

### **Prueba del teorema.**

Observando el código, podemos apreciar que el **peor caso** ocurre cuando el **arreglo** se encuentra ordenado de mayor a menor. En este caso, en cada pasada el elemento se movería terminaría en la primera casilla, por lo que se necesitaría realizar $n-1$ pasadas para que los elementos quedaran ordenados.

```python
arreglo =[5,4,3,2,1];
print("Entrada no ordenada:",arreglo)
contador = insertionSort(arreglo,True)
print("Salida ordenada:",arreglo)
print("Total de comparaciones realizadas:",contador)
```

En este caso, la primera pasada hizo 1 comparaciones, la segunda 2, la tercera 3 y la cuarta 4. Total: 10.

En el caso general, se deberían realizar $\sum\limits_{i=0}^{n-1} i=\frac{n(n-1)}{2}$ comparaciones. Por lo que el **tiempo de ejecución del algoritmo en el peor caso** es $O(n^2)$.

# **4. Correctitud**

### **Teorema (Correctitud).**

*El algoritmo **InsertionSort** genera un arreglo: $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir,* $a_1'\leq a_2' \leq... \leq a_n'$.

### **Prueba del Teorema**

Previamente observamos que en cada iteración de **InsertionSort** la cantidad de elementos ordenados aumenta agrupados en el lado izquierdo del arreglo.

Por lo que probaremos la siguiente **propiedad invariante de bucle** para el `Insertion Sort`:

> Al comienzo de cada iteración $i$ del bucle, los elementos previos del i-ésimo elemento del arreglo se encuentran ordenados de menor a mayor, y se deberá decidir donde insertar el i-ésimo elemento en los elementos ya ordenados.

**Inicialización**

Al comienzo de la primera iteración (i=1), los primeros elementos del arreglo  previos al elemento i-ésimo cumplen la propiedad. Esto es trivialmente cierto.

**Mantención**

Aplicando inducción, asumimos que al comienzo de cada iteración $i$ se cumple la propiedad. Es decir, los primeros elementos previos al i-ésimo del arreglo corresponden a elementos ordenados de menor a mayor.

Usemos como ejemplo el siguiente arreglo = **[2, 4, 3, 8, 1]**.

Se puede observar que los primeros 2 elementos se encuentran ya ordenados

Consideremos al $3$ como el i-ésimo elemento donde nos encontramos. Se puede observar que los 2 primeros elementos se encuentran ya ordenados

**[2, 4, $3$, 8, 1]**

Durante Insertion Sort, se puede apreciar que a la izquierda de $3$, el i-ésimo elemento, los números se encuentran ordenados de menor a mayor.

El objetivo sería agregar el $3$ al grupo de elementos ordenados del arreglo, primero se compara si $3$ < $4$. Como $3$ es menor, el $4$ ocupa la posición del $3$ y deja un espacio para el $3$

**[2, 4, 4, 8, 1]**

Ahora el $3$ lo comparamos con el $2$ (número previo al 4), como $3$ < $2$ no se cumple, el $2$ permanece en su lugar y el $3$ quedaría a la derecha del $2$.

**[2, $3$, 4, 8, 1]**

Ahora los 3 primeros elementos del arreglo se encuentran ordenados de menor a mayor. Al finalizar la iteración (y al comienzo de la siguiente), se seguirá cumpliendo la **propiedad invariante**. Es decir, los números a la izquierda del i-ésimo elemento ya se encontrarán ordenados y habrá que ingresar el i-ésimo elemento a los elementos ordenados

**Correctitud**

Finalmente, como la propiedad de bucle invariante es verdadera al inicio del bucle y se mantiene en cada iteración, podemos decir que al **finalizar la n-ésimo iteración del algoritmo**, se genera un arreglo $[a_1',a_2',...,a_n']$, con los mismos elementos del arreglo de entrada ordenados de menor a mayor.

## **5.1. Número de comparaciones** (Experimental)

A continuación, se muestra gráficamente una comparación entre:

* cantidad de comparaciones del peor caso para el algoritmo.
* cantidad de comparaciones del mejor caso para el algoritmo.

Los arreglos de entrada son generados aleatoriamente haciendo un sampling de un conjunto de números entre 1 y 100.

### Para tamaños de problemas con $n \in [4,30]$.

```python
import matplotlib.pyplot as plt

x=[n for n in range(4,30)] 
y1=[n*(n-1)/2 for n in range(4,30)] # worst case
y2=[n for n in range(4,30)] # best case
y=[]; 

for n in range(4,30):
    a = random.sample(range(1, 100), n)
    counter = insertionSort(a)
    y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor Caso Teóricamente", "Mejor Caso Teóricamente"])

## Nombre de los ejes
plt.xlabel('n')
plt.ylabel('Número de comparaciones')

# Mostrar Gráfico
plt.show()
```

### Para tamaños de problemas con $n \in [4,10]$.

```python
import matplotlib.pyplot as plt

x=[n for n in range(4,10)] 
y1=[n*(n-1)/2 for n in range(4,10)] # worst case
y2=[n for n in range(4,10)] # best case
y=[]; 

for n in range(4,10):
    a = random.sample(range(1, 100), n)
    counter = insertionSort(a)
    y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor Caso Teóricamente", "Mejor Caso Teóricamente"])

## Nombre de los ejes
plt.xlabel('n')
plt.ylabel('Número de comparaciones')

# Mostrar Gráfico
plt.show()
```

En el siguiente gráfico se reporta el tiempo de ejecución del algoritmo **InsertionSort** en función del tamaño del problema para arreglos generados aleatoriamente de tamaños entre $5$ y $600$.
